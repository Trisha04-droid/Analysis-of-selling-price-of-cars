```python
#import all the necessary libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import spicy as sp
```

```python
#read csv through pandas 
df=pd.read_csv('output.csv')
#print first 5 rows
df.head()
```

```python
#give headings to each of the columns to modify the data easily.
headers = ["symboling", "normalized-losses", "make",  
           "fuel-type", "aspiration","num-of-doors", 
           "body-style","drive-wheels", "engine-location", 
           "wheel-base","length", "width","height", "curb-weight", 
           "engine-type","num-of-cylinders", "engine-size",  
           "fuel-system","bore","stroke", "compression-ratio", 
           "horsepower", "peak-rpm","city-mpg","highway-mpg","price"] 
df.columns=headers
df.head()
```

```python
#find if there are any null or missing values
data = df
data.isna().any()
data.isnull().any()
```

```python
# converting mpg to L / 100km 
data['city-mpg']=235/df['city-mpg']
data.rename(columns={'city_mpg':"city-L/100km"},inplace=True)
print(data.columns)
#checking the data type of each column
data.dtypes
```

```python
#print all the unique values of price column
data.price.unique()
```

```python
#remove question marks from the column price
data = data[data.price != '?']
#convert it into integer type
data['price']=data['price'].astype(int)
data.dtypes
```

```python
#scaling the data to improve the score
data['length'] = data['length']/data['length'].max() 
data['width'] = data['width']/data['width'].max() 
data['height'] = data['height']/data['height'].max() 
```

```python
#to check prices of all the cars by grouping them
bins = np.linspace(min(data['price']), max(data['price']), 4)  
group_names = ['Low', 'Medium', 'High'] 
data['price-binned'] = pd.cut(data['price'], bins,  
                              labels = group_names,  
                              include_lowest = True)
```

```python
print(data['price-binned'])
```

```python
#plot a histogram
plt.hist(data['price-binned'])
plt.show()
```

```python
#converts categorical variables into numerical variables
pd.get_dummies(data['fuel-type'])
data.describe()
```

```python
#draw a box plot with the price data
plt.boxplot(data['price']) 
sns.boxplot(x ='drive-wheels', y ='price', data = data) #this is used for aesthetics 
```

```python
#draw a scatter plot between engine-size and price
plt.scatter(data['engine-size'], data['price']) 
plt.title('Scatterplot of Enginesize vs Price') 
plt.xlabel('Engine size') 
plt.ylabel('Price') 
plt.grid() 
plt.show() 
```

```python
#group the data by certain rows
test = data[['drive-wheels', 'body-style', 'price']] 
data_grp = test.groupby(['drive-wheels', 'body-style'],  
                         as_index = False).mean() 
  
data_grp 
```

```python
#this converts unique rows into columns
data_pivot = data_grp.pivot(index = 'drive-wheels', 
                            columns = 'body-style') 
data_pivot 
```

```python
#plot a heat map using data_pivot
plt.pcolor(data_pivot, cmap ='RdBu') 
plt.colorbar() 
plt.show()
```

```python
#draw a regplot using seaborn
sns.regplot(x ='engine-size', y ='price', data = data) 
plt.ylim(0, ) 
```

```python
#find statistic value and pvalue to understand whether it is okay or not
data_annova = data[['make','price']]
grouped_annova = data_annova.groupby(['make'])
annova_results_l = sp.stats.f_oneway( 
                             grouped_annova.get_group('honda')['price'], 
                             grouped_annova.get_group('subaru')['price'] 
                                    )
print(annova_results_l)
```
